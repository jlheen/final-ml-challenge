```python
import json
import requests
import pandas as pd
```

```python
apiKey = "ess-rymzVjj58kzq4Mzj"
ticker = 'BK'
url = f"https://www.quandl.com/api/v3/datasets/WIKI/{ticker}.json?start_date=2008-03-27&end_date=2018-03-27&api_key={apiKey}"
```

```python
response = requests.get(url).json()
```

```python
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
date = []
open_amt = []
high = []
low = []
close = []
volume = []
dividend = []
split = []
adj_open = []
adj_high = []
adj_low = []
adj_close = []
adj_volume = []
```

```python
daily_ticks = len(response['dataset']['data'])
for x in range(daily_ticks):
    try:
        date.append(response['dataset']['data'][x][0])
    except: 
        date.append('')
    try:
        open_amt.append(response['dataset']['data'][x][1])
    except: 
        open_amt.append('')
    try:
        high.append(response['dataset']['data'][x][2])
    except: 
        high.append('')
    try:
        low.append(response['dataset']['data'][x][3])
    except: 
        low.append('')
    try:
        close.append(response['dataset']['data'][x][4])
    except: 
        close.append('')
    try:
        volume.append(response['dataset']['data'][x][5])
    except: 
        volume.append('')
    try:
        dividend.append(response['dataset']['data'][x][6])
    except: 
        dividend.append('')
    try:
        split.append(response['dataset']['data'][x][7])
    except: 
        split.append('')
    try:
        adj_open.append(response['dataset']['data'][x][8])
    except: 
        adj_open.append('')
    try:
        adj_high.append(response['dataset']['data'][x][9])
    except: 
        adj_high.append('')
    try:
        adj_low.append(response['dataset']['data'][x][10])
    except: 
        adj_low.append('')
    try:
        adj_close.append(response['dataset']['data'][x][11])
    except: 
        adj_close.append('')
    try:
        adj_volume.append(response['dataset']['data'][x][12])
    except: 
        adj_volume.append('')
```

```python
df = pd.DataFrame({
    'ticker': ticker,
    'date': date,
    'open': open_amt,
    'high': high,
    'low': low,
    'close': close,
    'volume': volume,
    'dividend': dividend,
    'split': split,
    'adj_open': adj_open,
    'adj_high': adj_high,
    'adj_low': adj_low,
    'adj_close': adj_close,
    'adj_volume': adj_volume
})
```

```python
df.to_csv(f"data/{ticker}.csv", index=False, header=True)
```
